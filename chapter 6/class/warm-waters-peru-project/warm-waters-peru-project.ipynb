Centuries ago, Peruvian fishing communities coined the term El Niño to refer to years when warm ocean temperatures would affect fishing productivity.

In this project we will investigate the relationship between El Niño events and word usage levels for different species of fish.

### Fish words and El Niño
We will use the `plot_animal()` function we developed in the earlier project:

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 10)

# ── Load both datasets ─────────────────────────────────────────────────────────
word_trends = pd.read_csv('animal-word-trends-peru.csv')
catch_size  = pd.read_csv('anchoveta-caught-per-year.csv')

EL_NINO_YEARS = [1965, 1973, 1983, 1987]

# ── Helper function ────────────────────────────────────────────────────────────
def plot_peru_animal(animal):
    plt.figure(figsize=(8, 3))
    trend = word_trends.query("word == @animal")
    plt.plot(trend['year'], trend['frequency'], label=animal, linewidth=2)

    for i, year in enumerate(EL_NINO_YEARS):
        plt.axvline(year, label='El Niño event' if i == 0 else '',
                    color='magenta', ls='--')

    plt.ylabel('Frequency per million')
    plt.title(f'El Niño events and word frequency for "{animal}"')
    plt.xlim(1960, 1990)
    plt.ylim(0, None)
    plt.legend()
    plt.gca().spines[['top', 'right']].set_visible(False)
    plt.tight_layout()
    plt.show()
```

### Multiple hypotheses
The word "anchoveta" appears to show an increased frequency of usage around large El Niño years. Let's consider a variety of hypotheses for why this might be the case. Here are a few possibilities:

- Abundance hypothesis: Perhaps anchoveta thrive in warm water and are more abundant in El Niño years. This could cause people to mention anchoveta more.

- Scarcity hypothesis: Perhaps anchoveta populations decline in warm water. This could cause the fishing community to be concerned and therefore discuss anchoveta more.

- Exception hypothesis: Perhaps most fish populations are affected by El Niño with anchoveta being a rare exception. This could cause an increased interest in anchoveta, such as research into why anchoveta is not affected like the other fish.

For each El Niño event, the frequency of the word "anchoveta" increases while the amount of anchoveta caught decreases.

```python
# ── 3. Catch size line plot with El Niño markers ──────────────────────────────
plt.figure(figsize=(8, 3))
plt.plot(catch_size['year'], catch_size['megatonnes'], color='green', linewidth=2)

for i, year in enumerate(EL_NINO_YEARS):
    plt.axvline(year, label='El Niño event' if i == 0 else '',
                color='magenta', ls='--')

plt.ylabel('Megatonnes')
plt.title('Megatonnes of Anchoveta Caught per Year')
plt.legend()
ax = plt.gca()
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# ── 4. Combined chart: word frequency + catch size side by side ───────────────
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Top: anchoveta word frequency
trend = word_trends.query("word == 'anchoveta'")
axes[0].plot(trend['year'], trend['frequency'],
             color='steelblue', linewidth=2, label='Word frequency')
for i, year in enumerate(EL_NINO_YEARS):
    axes[0].axvline(year, color='magenta', ls='--',
                    label='El Niño event' if i == 0 else '')
axes[0].set_ylabel('Frequency per million')
axes[0].set_title('Anchoveta Word Frequency vs Catch Size — El Niño Inverse Relationship')
axes[0].legend(fontsize=9)
axes[0].set_xlim(1960, 1990)
axes[0].set_ylim(0, None)
axes[0].spines[['top', 'right']].set_visible(False)

# Bottom: catch size
axes[1].plot(catch_size['year'], catch_size['megatonnes'],
             color='green', linewidth=2, label='Catch size')
for i, year in enumerate(EL_NINO_YEARS):
    axes[1].axvline(year, color='magenta', ls='--',
                    label='El Niño event' if i == 0 else '')
axes[1].set_ylabel('Megatonnes caught')
axes[1].set_xlabel('Year')
axes[1].legend(fontsize=9)
axes[1].set_xlim(1960, 1990)
axes[1].set_ylim(0, None)
axes[1].spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# ── 5. All five fish on one chart for comparison ──────────────────────────────
plt.figure(figsize=(10, 5))

colors = ['#E53935', '#FB8C00', '#43A047', '#1565C0', '#6A1B9A']
for fish, color in zip(fish_words, colors):
    trend = word_trends.query("word == @fish")
    plt.plot(trend['year'], trend['frequency'],
             label=fish.title(), color=color, linewidth=2)

for i, year in enumerate(EL_NINO_YEARS):
    plt.axvline(year, color='magenta', ls='--', linewidth=1.2,
                label='El Niño event' if i == 0 else '')

plt.ylabel('Frequency per million')
plt.xlabel('Year')
plt.title('All Peruvian Fish Words vs El Niño Events (1960–1990)')
plt.xlim(1960, 1990)
plt.ylim(0, None)
plt.legend(fontsize=9)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# ── 6. Before vs during El Niño: anchoveta word frequency ─────────────────────
print("Anchoveta Word Frequency: Normal Years vs El Niño Years")
print("=" * 50)

anch = word_trends.query("word == 'anchoveta'")

normal_freq  = anch[~anch['year'].isin(EL_NINO_YEARS)]['frequency'].mean()
elnino_freq  = anch[anch['year'].isin(EL_NINO_YEARS)]['frequency'].mean()
change       = (elnino_freq - normal_freq) / normal_freq * 100

print(f"Avg frequency — Normal years : {normal_freq:.4f} per million")
print(f"Avg frequency — El Niño years: {elnino_freq:.4f} per million")
print(f"Increase during El Niño      : +{change:.1f}%")

print()
print("Anchoveta Catch: Normal Years vs El Niño Years")
print("=" * 50)

normal_catch = catch_size[~catch_size['year'].isin(EL_NINO_YEARS)]['megatonnes'].mean()
elnino_catch = catch_size[catch_size['year'].isin(EL_NINO_YEARS)]['megatonnes'].mean()
catch_change = (elnino_catch - normal_catch) / normal_catch * 100

print(f"Avg catch — Normal years : {normal_catch:.2f} megatonnes")
print(f"Avg catch — El Niño years: {elnino_catch:.2f} megatonnes")
print(f"Change during El Niño    : {catch_change:.1f}%")
```
