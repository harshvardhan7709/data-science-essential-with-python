# Naming colors

If you are shown random colors evenly spaced across the color spectrum, what percentage of those colors would the average person name red? What about green? 

English speakers in the United States were shown 80 color chips in random order in controlled lighting condtions. These 80 colors are evenly spaced across the standard Munsell array of colors. The participants were told: "There are 11 choices: black, white, red, green, blue, purple, brown, yellow, orange, pink, gray. Choose the closest color word".

This study was then repeated with Bolivian-Spanish speakers in Bolivia and Tsimane' speakers from the Amazon.

The dataset `munsell-array-fixed-choice.csv` shows the most common name in each language for each color chip.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("munsell-array-fixed-choice.csv")
df.head()
```

The function below plots the 80 evenly spaced colors that each study participant was shown in random order.

```python
def plot_colors(color_column):
    fig, ax = plt.subplots()
    for idx, row in df.iterrows():
        ax.add_patch(plt.Rectangle((row['x'] - 0.5, row['y'] - 0.5), 1, 1, color=row[color_column], linewidth=0))

    ax.set_xlim(0.5, 20.5)
    ax.set_ylim(0.5, 8.5)
    ax.set_aspect('equal')
    ax.set_facecolor('black')
    ax.set_xticks(range(1, 21))
    ax.set_yticks(range(1, 9), ['H', 'G', 'F', 'E', 'D', 'C', 'B', 'A'])

plot_colors('tile_hex')
```

We can also use this function to plot the most common name in each language for each color chip.

```python
# Also try 'spanish_color' and 'tsimane_color'
plot_colors('english_color')
```

```python
english_counts = df[
    "english_color"
].value_counts().reset_index()
```

```python
english_counts.columns = [
    "color",
    "count"
]
```

```python
english_counts["percentage"] = (
    english_counts["count"] /
    english_counts["count"].sum()
) * 100
```

```python
english_sorted = english_counts.sort_values(
    "percentage"
)

plt.figure(figsize=(10,6))

plt.barh(
    english_sorted["color"],
    english_sorted["percentage"]
)

plt.xlabel("Percentage")
plt.ylabel("Color")

plt.title(
    "English Color Naming Distribution"
)

plt.tight_layout()

plt.show()
```

```python
spanish_counts = df[
    "spanish_color"
].value_counts().reset_index()

spanish_counts.columns = [
    "color",
    "count"
]

spanish_counts["percentage"] = (
    spanish_counts["count"] /
    spanish_counts["count"].sum()
) * 100
```

```python
spanish_sorted = spanish_counts.sort_values(
    "percentage"
)

plt.figure(figsize=(10,6))

plt.barh(
    spanish_sorted["color"],
    spanish_sorted["percentage"]
)

plt.title(
    "Spanish Color Naming Distribution"
)

plt.tight_layout()

plt.show()
```

```python
tsimane_counts = df[
    "tsimane_color"
].value_counts().reset_index()

tsimane_counts.columns = [
    "color",
    "count"
]

tsimane_counts["percentage"] = (
    tsimane_counts["count"] /
    tsimane_counts["count"].sum()
) * 100
```

```python
tsimane_sorted = tsimane_counts.sort_values(
    "percentage"
)

plt.figure(figsize=(10,6))

plt.barh(
    tsimane_sorted["color"],
    tsimane_sorted["percentage"]
)

plt.title(
    "Tsimane Color Naming Distribution"
)

plt.tight_layout()

plt.show()
```

```python
english_compare = english_counts[
    ["color", "percentage"]
].rename(
    columns={
        "percentage": "english_percent"
    }
)

tsimane_compare = tsimane_counts[
    ["color", "percentage"]
].rename(
    columns={
        "percentage": "tsimane_percent"
    }
)
```

```python
comparison = pd.merge(
    english_compare,
    tsimane_compare,
    on="color",
    how="inner"
)
```

```python
plt.figure(figsize=(8,8))

plt.scatter(
    comparison["english_percent"],
    comparison["tsimane_percent"]
)

plt.xlabel("English Percentage")
plt.ylabel("Tsimane Percentage")

plt.title(
    "English vs Tsimane Color Usage"
)

plt.grid(alpha=0.3)

plt.show()
```

```python
correlation = comparison[
    ["english_percent", "tsimane_percent"]
].corr()

print(correlation)
```
